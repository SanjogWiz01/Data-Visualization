```python
import pandas as pd
import numpy as np
```

```python
## Creating a Pandas Series
```

```python
data_list = [10, 20, 30, 40, 50]
series_from_list = pd.Series(data_list)
series_from_list
```

```python
custom_index = ['a', 'b', 'c', 'd', 'e']
series_custom_index = pd.Series(data_list, index=custom_index)
series_custom_index
```

```python
array_data = np.array([100, 200, 300, 400, 500])
series_from_array = pd.Series(array_data, index=['p', 'q', 'r', 's', 't'])
series_from_array
```

```python
s1 = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
s2 = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'])

addition = s1 + s2
subtraction = s2 - s1
multiplication = s1 * s2
division = s2 / s1

addition, subtraction, multiplication, division
```

```python
# mini work here
students = ['Ram', 'Sita', 'Bishal', 'Nisha', 'Hari']
marks = [78, 85, 67, 90, 74]

marks_series = pd.Series(marks, index=students)
marks_series
print(marks_series['Sita'])
print(marks_series[['Ram', 'Nisha']])
```

```python
print("Marks of Sita:", marks_series['Sita'])
print("Marks of Nisha:", marks_series['Nisha'])
```

```python
bonus = pd.Series([2, 2, 2, 2, 2], index=students)
final_marks = marks_series + bonus
final_marks
```

```python
print("Highest Marks:", final_marks.max())
print("Lowest Marks:", final_marks.min())
```

```python
result = final_marks >= 75
result
```

```python
print("Average Marks:", final_marks.mean())
print("Total Students:", final_marks.count())
```
