```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('collection_data.csv')
print(df) # orginal data tracking there is nan value also
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
df
```

```python
# Fill missing temperature with mean temperature
df["temperature"] = df["temperature"].fillna(df["temperature"].mean())

print("\nAfter cleaning:")
print(df)
```

```python
# now analizing the real data using libary
# Hottest and coldest day
hottest_day = df.loc[df["temperature"].idxmax()]
coldest_day = df.loc[df["temperature"].idxmin()]

print("\nHottest day:")
print(hottest_day)

print("\nColdest day:")
print(coldest_day)
```

```python
temps = df["temperature"].to_numpy()  # using this function to find the stasts using numpy

print("\nTemperature stats:")
print("Mean:", np.mean(temps))
print("Max:", np.max(temps))
print("Min:", np.min(temps))
```

```python
plt.figure()
plt.plot(df["date"], df["temperature"])
plt.title("Temperature Over Time")
plt.xlabel("Date")
plt.ylabel("Temperature (Â°C)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
